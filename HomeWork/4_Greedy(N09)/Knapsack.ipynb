```python
n=int(input())
value=list(map(int,input().split()))
weight=list(map(int,input().split()))
m=int(input())
def swap(i,j):
  tmp=value[i]
  value[i]=value[j]
  value[j]=tmp
  tmp=weight[i]
  weight[i]=weight[j]
  weight[j]=tmp
def bubble_sort(a,b,n):
  for i in range(n-1):
    for j in range(i+1,n):
      if (a[i]/b[i])<(a[j]/b[i]):
        swap(i,j)
bubble_sort(value,weight,n)
remain=m
res=0
for i in range(n):
  if (weight[i]<=remain):
    remain-=weight[i]
    res+=value[i]
print(res)
```
